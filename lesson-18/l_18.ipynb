```python
import pandas as pd
df = pd.read_csv('customer_orders.csv')
df
```

```python
#1
x = df.groupby('CustomerID').agg({'Quantity' : 'sum'})
y = x['Quantity'] < 26
s = x[y]
s
```

```python
#2
x = df.groupby(['CustomerID', 'Product']).agg({'Price' : 'mean'})
y = x['Price'] > 120
s = x[y]
s
```

```python
#3
x = df.groupby('Product')[['Quantity', 'Price']].agg('sum')
y = x['Quantity'] < 5
s = x[y]
s
```
